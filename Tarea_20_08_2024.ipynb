Resolviendo $A \cdot \text{rect}_\tau(t)$ utilizando la Densidad Espectral de Potencia (PSD) y aplicando el método de Welch con una ventana de Hanning

1. **Definir la señal:**
La función rectángulo $\text{rect}_\tau(t)$ es una señal que es 1 dentro del intervalo $[-\tau/2, \tau/2]$ y 0 fuera de este. Multiplicada por una constante $A$, tendríamos:

$$
x(t) = A \cdot \text{rect}_\tau(t)
$$

donde $A$ puede ser un valor aleatorio.

2. **Generar la señal en el dominio del tiempo:**
Simulamos la señal $x(t)$ en un intervalo de tiempo discreto.

3. **Aplicar el método de Welch:**
Utilizamos el método de Welch para estimar la PSD de la señal $x(t)$, utilizando una ventana de Hanning.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import welch
from scipy.signal.windows import hann

# Parámetros
A = 1  # Valor aleatorio para A
tau = 1  # Duración del pulso
fs = 1000  # Frecuencia de muestreo
t = np.arange(-2*tau, 2*tau, 1/fs)  # Tiempo

# Definir la señal rectángulo
x_t = A * np.where(np.abs(t) <= tau/2, 1, 0)

# Aplicar el método de Welch con ventana de Hanning
window = hann(len(x_t))
f, Pxx = welch(x_t, fs, window=window, nperseg=len(x_t))

# Graficar la Densidad Espectral de Potencia (PSD)
plt.figure(figsize=(10, 6))
plt.semilogy(f, Pxx)
plt.title('Densidad Espectral de Potencia (PSD) usando el Método de Welch')
plt.xlabel('Frecuencia [Hz]')
plt.ylabel('PSD [V^2/Hz]')
plt.grid(True)
plt.show()
```
